## 1. Introdução ao SVM
### Introdução ao Support Vector Machines (SVM)

Support Vector Machines (SVM) são uma poderosa ferramenta de aprendizado de máquina usada tanto para classificação quanto para regressão. A ideia principal do SVM é encontrar um hiperplano em um espaço de alta dimensão que separa diferentes classes. Para a regressão, a ideia é encontrar um hiperplano que esteja o mais próximo possível dos dados.

Os principais conceitos por trás do SVM são:
- Margem: A distância entre o hiperplano de separação e os pontos de dados mais próximos de qualquer classe.
- Vetores de Suporte: Os pontos de dados que estão mais próximos do hiperplano e influenciam a posição e orientação do hiperplano.
- Kernel: Uma função que transforma os dados em um espaço de alta dimensão, permitindo que o SVM resolva problemas não linearmente separáveis.


## 2. Explicação dos parâmetros do SVM do Scikit-learn

### Parâmetros do SVM no Scikit-learn

O Scikit-learn fornece várias opções para configurar o SVM. Alguns dos principais parâmetros são:

- `C`: Parâmetro de regularização. Quanto maior o valor de `C`, menor a margem de erro, mas pode levar ao overfitting.
- `kernel`: Especifica o tipo de kernel a ser usado na função de decisão. Pode ser 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed' ou um callable.
- `degree`: Grau do polinômio se o kernel for 'poly'. Ignorado por outros kernels.
- `gamma`: Coeficiente do kernel para 'rbf', 'poly' e 'sigmoid'. Se `gamma` for 'scale', `gamma` será 1 / (n_features * X.var()), se for 'auto', `gamma` será 1 / n_features.
- `coef0`: Término independente no kernel 'poly' e 'sigmoid'.
- `shrinking`: Se deve usar heurísticas de encolhimento.
- `probability`: Se deve habilitar estimativas de probabilidade. Isso é mais demorado e deve ser ativado antes de ajustar o modelo.
- `tol`: Critério de tolerância para parada.
- `cache_size`: Tamanho do cache em MB.
- `class_weight`: Peso associado às classes.
- `verbose`: Habilitar saída detalhada.
- `max_iter`: Número máximo de iterações.

```python
import matplotlib.pyplot as plt

from sklearn import datasets, svm
from sklearn.inspection import DecisionBoundaryDisplay

# import some data to play with
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

## 4. Implementação do SVM para regressão

```python
#Vamos usar o dataset California Housing da UCI para um exemplo de regressão.
# Importar bibliotecas necessárias
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
# Carregar dataset de exemplo
california = fetch_california_housing()
X = california.data
y = california.target

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar o modelo SVM de regressão
reg = SVR(C=1.0, kernel='rbf', gamma='scale')

# Treinar o modelo
reg.fit(X_train, y_train)

# Fazer previsões
y_pred = reg.predict(X_test)

# Avaliar o modelo
mse = mean_squared_error(y_test, y_pred)
print(f'Erro quadrático médio da regressão: {mse}')

# Visualização
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.title('SVM Regression Results')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

# Calcular o MAE
mae = mean_absolute_error(y_test, y_pred)
print(f'Erro absoluto médio da regressão: {mae}')

# Visualização detalhada
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.title('SVM Regression Results')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.svm import SVR
```

```python
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

# add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
svr_rbf = SVR(kernel="rbf", C=100, gamma=0.1, epsilon=0.1)
svr_lin = SVR(kernel="linear", C=100, gamma="auto")
svr_poly = SVR(kernel="poly", C=100, gamma="auto", degree=3, epsilon=0.1, coef0=1)
```

```python
lw = 2

svrs = [svr_rbf, svr_lin, svr_poly]
kernel_label = ["RBF", "Linear", "Polynomial"]
model_color = ["m", "c", "g"]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 10), sharey=True)
for ix, svr in enumerate(svrs):
    axes[ix].plot(
        X,
        svr.fit(X, y).predict(X),
        color=model_color[ix],
        lw=lw,
        label="{} model".format(kernel_label[ix]),
    )
    axes[ix].scatter(
        X[svr.support_],
        y[svr.support_],
        facecolor="none",
        edgecolor=model_color[ix],
        s=50,
        label="{} support vectors".format(kernel_label[ix]),
    )
    axes[ix].scatter(
        X[np.setdiff1d(np.arange(len(X)), svr.support_)],
        y[np.setdiff1d(np.arange(len(X)), svr.support_)],
        facecolor="none",
        edgecolor="k",
        s=50,
        label="other training data",
    )
    axes[ix].legend(
        loc="upper center",
        bbox_to_anchor=(0.5, 1.1),
        ncol=1,
        fancybox=True,
        shadow=True,
    )

fig.text(0.5, 0.04, "data", ha="center", va="center")
fig.text(0.06, 0.5, "target", ha="center", va="center", rotation="vertical")
fig.suptitle("Support Vector Regression", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.inspection import DecisionBoundaryDisplay

# we create 40 separable points
X, y = make_blobs(n_samples=40, centers=2, random_state=6)

# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel="linear", C=1000)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
