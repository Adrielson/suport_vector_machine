```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Carregando o conjunto de dados Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Dividindo o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Padronizando os dados
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Função para treinar e avaliar o SVM com diferentes kernels
def train_and_evaluate_svm(kernel):
    svm = SVC(kernel=kernel)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    
    print(f"Kernel: {kernel}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("\n")

# Treinando e avaliando o SVM com diferentes kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

for kernel in kernels:
    train_and_evaluate_svm(kernel)

# Visualizando os resultados
results = []

for kernel in kernels:
    svm = SVC(kernel=kernel)
    svm.fit(X_train, y_train)
    score = svm.score(X_test, y_test)
    results.append((kernel, score))

results_df = pd.DataFrame(results, columns=['Kernel', 'Accuracy'])
results_df.plot(kind='bar', x='Kernel', y='Accuracy', legend=False, title='SVM Kernel Comparison')
plt.ylabel('Accuracy')
plt.show()
```

```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Carregando o conjunto de dados Breast Cancer Wisconsin
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

```

```python
# Dividindo o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Padronizando os dados
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Função para treinar e avaliar o SVM com diferentes kernels
def train_and_evaluate_svm(kernel):
    svm = SVC(kernel=kernel)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    
    print(f"Kernel: {kernel}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Treinando e avaliando o SVM com diferentes kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
accuracies = []

for kernel in kernels:
    accuracy = train_and_evaluate_svm(kernel)
    accuracies.append((kernel, accuracy))

# Visualizando os resultados
results_df = pd.DataFrame(accuracies, columns=['Kernel', 'Accuracy'])
plt.figure(figsize=(10, 6))
sns.barplot(x='Kernel', y='Accuracy', data=results_df)
plt.title('SVM Kernel Comparison')
plt.xlabel('Kernel')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.show()
```
